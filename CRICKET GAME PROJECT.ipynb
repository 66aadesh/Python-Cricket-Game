```python

# Importing the necessary module
import random

# List to store possible runs
runs_list = [0, 1, 2, 4, 6]  # 0=no run, 1=1 run, 2=2 runs, 4=4 runs, 6=6 runs

# Function to randomly select a run from runs_list
def runs_scored():
    ''' Function to pull random scores from runs_list '''
    run_scored = random.choice(runs_list)
    return run_scored

# List to store bowling outcomes: 0=score, 1=wicket
bowling_outcome_list = [0, 1]

# Probability of scoring a run (0) or getting a wicket (1)
bowling_outcome_probability = [0.84, 0.16]  # Higher probability for scoring a run

# Function to randomly select a bowling outcome based on probability
def bowl_outcome():
    ''' Function to pull random scores from bowling_outcome_list '''
    bowling_outcome = random.choices(bowling_outcome_list, bowling_outcome_probability, k=1)[0]
    return bowling_outcome

# Function to simulate throwing the ball for an over (6 balls)
def throw_ball():
    ''' Ask the user to throw a ball and set the game in motion for 1 over (6 balls) '''
    latest_r = 0  # Variable to track latest runs scored
      
    # Loop for 6 balls in an over
    for j in range(1, 7):
        bowl_throw = input("Press 'Y' to throw {} ball".format(j))
        if bowl_throw != 'Y':
            print("You have forfeited the game.")
            break
        
        # Determine outcome: score or wicket
        b = bowl_outcome()
        if b == 0:
            r = runs_scored()
            print("You scored {} runs in {} ball.".format(r, j))
        else:
            print("You are out.")
            break
        
        latest_r += r  # Accumulating the runs scored
    
    total_score = latest_r
    print("Your total Score was {}".format(total_score))
    return total_score

# List for the toss: H=Heads & T=Tails
toss_list = ['H', 'T']

# Function to simulate a toss
def toss():
    ''' Function to draw a toss to select who will bat first '''
    toss_result = random.choice(toss_list)
    return toss_result

# Function to start the match
def match_start():
    ''' Use all the functions to start the entire match '''
    # Getting player names
    player1 = input("Please enter Player1 name: -")
    player2 = input("Please enter Player2 name: -")
    
    # Performing the toss
    t = toss()
    
    if t == 'H':  # Player1 wins the toss with Heads
        player1_choice = input("{} please choose to 'BAT' or 'BOWL'.".format(player1))

        if player1_choice == 'BAT':
            # Player1 is batting & player2 is bowling
            print("{} has chosen to BAT".format(player1))
            print("FIRST INNINGS - {}".format(player1))
            first_innings_score = throw_ball()
            print("SECOND INNINGS - {}".format(player2))
            second_innings_score = throw_ball()
        
            # Determining the winner based on scores
            if first_innings_score > second_innings_score:
                print("{} has won!".format(player1))
            else:
                print("{} has won!".format(player2))
        else:
            # Player2 is batting & player1 is bowling
            print("{} has chosen to BOWL".format(player1))
            print("FIRST INNINGS - {}".format(player2))
            first_innings_score = throw_ball()
            print("SECOND INNINGS - {}".format(player1))
            second_innings_score = throw_ball()
        
            # Determining the winner based on scores
            if first_innings_score > second_innings_score:
                print("{} has won!".format(player2))
            else:
                print("{} has won!".format(player1))
    else:
        player2_choice = input("{} please choose to 'BAT' or 'BOWL'.".format(player2))
        if player2_choice == 'BAT':
            # Player2 is batting & player1 is bowling
            print("{} has chosen to BAT".format(player2))
            print("FIRST INNINGS - {}".format(player2))
            first_innings_score = throw_ball()
            print("SECOND INNINGS - {}".format(player1))
            second_innings_score = throw_ball()
        
            # Determining the winner based on scores
            if first_innings_score > second_innings_score:
                print("{} has won!".format(player2))
            else:
                print("{} has won!".format(player1))
        else:
            # Player1 is batting & player2 is bowling
            print("{} has chosen to BOWL".format(player2))
            print("FIRST INNINGS - {}".format(player1))
            first_innings_score = throw_ball()
            print("SECOND INNINGS - {}".format(player2))
            second_innings_score = throw_ball()
        
            # Determining the winner based on scores
            if first_innings_score > second_innings_score:
                print("{} has won!".format(player1))
            else:
                print("{} has won!".format(player2))

# Starting the match
match_start()
```
